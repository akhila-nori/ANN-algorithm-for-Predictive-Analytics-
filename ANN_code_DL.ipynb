```python
X = df.iloc[ : , 1 : -1] # EXTRACTING X .... All rows and col all excluding col 0 and last column

y = df.iloc[ : , -1] #Printing all rows from -1 column i.e., last column
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
#MinMax scaling
from sklearn.preprocessing import MinMaxScaler

#creating object of MinMax Scaler Class

scaler = MinMaxScaler()

X_scaled = scaler.fit_transform(X)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

```

```python
#X_train_scaled #numpy array in which X-trained values are sclaed to a similar to values
```

```python
#Building neural network architecture, for that importing all the libraries

import tensorflow 
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense

#print(keras.__version__)
```

```python
#ANN model
model = Sequential() #model is Sequential class ka object
model.add(Dense(16, activation = 'relu', input_dim=16)) #16 layers 
model.add(Dense(1, activation = 'linear')) #Output layer = 1 
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error', optimizer = 'Adam')
```

```python
history = model.fit(X_train_scaled,y_train,epochs=20,validation_split=0.2) #model is trained here
#history.history['accuracy']
```

```python
y_pred = model.predict(X_test_scaled)
#y_pred = model.predict(X_scaled)
#X
print(y_pred)
#len(y_pred)
len(y_test)
```
